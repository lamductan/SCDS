```python
fig, ax = plt.subplots(1,1)
ax.plot(r_s, np.ones(len(r_s)), marker='v', markersize=markersize, label="beta omni")
ax.plot(r_s, df["beta_Aschner"], marker='D', markersize=markersize, label="beta Aschner")
ax.plot(r_s, df["beta_Tran"], marker='^', markersize=markersize, label="beta Tran")
ax.plot(r_s, df["beta_Lam_LEF"], marker='o', markersize=markersize, label="beta LEF")
ax.plot(r_s, df["beta_Lam_SEF"], marker='x', markersize=markersize, label="beta SEF")
ax.set_xticks(r_s)
ax.set_yticks(np.arange(1, 6.5, 0.5))

ax2 = ax.twinx()
w = 5
alpha=0.3
ax2.bar(r_s, df["n_total_omni"], width=w, label="total omni nodes", alpha=alpha)
ax2.bar(r_s, df["n_total_Aschner"], width=w, label="total nodes of Aschner", alpha=alpha)
ax2.bar(r_s, df["n_total_Tran"], width=w, label="total nodes of Tran", alpha=alpha)
ax2.bar(r_s, df["n_total_Lam_LEF"], width=w, label="total nodes of LEF", alpha=alpha)
ax2.bar(r_s, df["n_total_Lam_SEF"], width=w, label="total nodes of SEF", alpha=alpha)
ax2.set_yticks([100, 200, 400, 600, 800, 1000, 1200])
ax.set_title("(a)", y=-0.25)
#ax.legend(loc='upper right')
plt.savefig("img/download.png")
```

```python
def plot_df_1(df, ax, x_label, title, y_title=-0.4):
    r = df[x_label]
    ax.plot(r, df["n_total_omni"], marker='v', markersize=markersize, label="total omni nodes")
    ax.plot(r, df["n_total_Aschner"], marker='D', markersize=markersize, label="total nodes of Aschner")
    ax.plot(r, df["n_total_Tran"], marker='^', markersize=markersize, label="total nodes of Tran")
    ax.plot(r, df["n_total_Lam_LEF"], marker='o', markersize=markersize, label="total nodes of LEF")
    ax.plot(r, df["n_total_Lam_SEF"], marker='x', markersize=markersize, label="total nodes of SEF")
    ax.set_yticks([200, 400, 600, 800, 1000, 1200])
    ax.set_ylabel("Number of nodes used")
    ax.set_xticks(r)
    ax.set_xlabel(x_label)
    ax.set_title(title, y=y_title)
    return ax
```

```python
def plot_df_2(df, ax, x_label, title, y_title=-0.4):
    r = df[x_label]
    ax.plot(r, np.ones(len(r_s)), marker='v', markersize=markersize, label=r'$\beta$ of Han')
    ax.plot(r, df["beta_Aschner"], marker='D', markersize=markersize, label=r'$\beta$ of Aschner')
    ax.plot(r, df["beta_Tran"], marker='^', markersize=markersize, label=r'$\beta$ of Tran')
    ax.plot(r, df["beta_Lam_LEF"], marker='o', markersize=markersize, label=r'$\beta$ of LEF')
    ax.plot(r, df["beta_Lam_SEF"], marker='x', markersize=markersize, label=r'$\beta$ of SEF')
    ax.set_yticks(np.arange(1, 7, 1))
    ax.set_ylabel(r'$\beta$')
    ax.set_xticks(r)
    ax.set_xlabel(x_label)
    ax.set_title(title, y=y_title)
    return ax
```

```python
def plot_df_3(df, ax, x_label, title, y_title=-0.4):
    r = df[x_label]
    ax.plot(r, df["n_total_omni"], marker='v', markersize=markersize, label="total omni nodes")
    ax.plot(r, df["n_total_Aschner"], marker='D', markersize=markersize, label="total nodes of Aschner")
    ax.plot(r, df["n_total_Tran"], marker='^', markersize=markersize, label="total nodes of Tran")
    ax.plot(r, df["n_total_Lam_LEF"], marker='o', markersize=markersize, label="total nodes of LEF")
    ax.plot(r, df["n_total_Lam_SEF"], marker='x', markersize=markersize, label="total nodes of SEF")
    ax.set_yticks([1000, 2000, 3000, 4000, 5000, 6000])
    ax.set_ylabel("Number of nodes used")
    ax.set_xticks(r)
    ax.set_xlabel(x_label)
    ax.set_title(title, y=y_title)
    return ax
```

```python
fig, axes = plt.subplots(1, 3, figsize=(30, 5))
plot_df_1(df_fixed_rc_60, axes[0], "r_s", r"(a) $\theta_s=\pi/3$", -0.32)
plot_df_1(df_fixed_rc_90, axes[1], "r_s", r"(b) $\theta_s=\pi/2$", -0.32)
plot_df_1(df_fixed_rc_120, axes[2], "r_s", r"(c) $\theta_s=2\pi/3$", -0.32)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, bbox_to_anchor=(0.77, 1.2), ncol=5)
fig.savefig("img/total_nodes_fixed_rc.pdf", bbox_inches="tight")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(24, 4))
plot_df_2(df_fixed_rc_60, axes[0], "r_s", r"(a) $\theta_s=\pi/3$", -0.4)
plot_df_2(df_fixed_rc_90, axes[1], "r_s", r"(b) $\theta_s=\pi/2$", -0.4)
plot_df_2(df_fixed_rc_120, axes[2], "r_s", r"(c) $\theta_s=2\pi/3$", -0.4)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, bbox_to_anchor=(0.7, 1.3), ncol=5)
fig.savefig("img/beta_fixed_rc.pdf", bbox_inches="tight")
```
